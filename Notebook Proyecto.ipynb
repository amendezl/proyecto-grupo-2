```python
# Se definen listas de columnas de procedimientos y diagnósticos

procedimientos = [f'PROCEDIMIENTO{i}' for i in range(1, 11)]
diagnosticos   = [f'DIAGNOSTICO{i}'  for i in range(1, 11)]

# Se asegura de que cada código está como string sin espacios ni comas

for col in procedimientos:
    df_todos[col] = df_todos[col]\
        .astype(str)\
        .str.replace(',', '.', regex=False)\
        .str.strip()
```

```python
# Se define el set de códigos de trasplante renal

codigos_trasplante = {'55.6', '55.61', '55.69'}

# Se construye la máscara: True si alguna de las 10 columnas coincide
mask_trasplante = df_todos[procedimientos] \
    .isin(codigos_trasplante) \
    .any(axis=1)

# Se aplica el filtro
df_trasplantes = df_todos[mask_trasplante].copy()

# Verificar
print("Registros con trasplante renal:", df_trasplantes.shape[0])

df_trasplantes.head()
```

```python
print("Pacientes con transplante renal:", len(df_trasplantes["CIP_ENCRIPTADO"].unique())) #Cantidad de pacientes
```

```python
df_trasplantes[df_trasplantes.duplicated(subset='CIP_ENCRIPTADO', keep=False)] #Muestra los unicos dos pacientes que tienen dos registros
```

```python
# Se Calcula la edad en años a la fecha de ingreso

df_trasplantes['EDAD'] = (
    df_trasplantes['FECHA_INGRESO'] - df_trasplantes['FECHA_NACIMIENTO']
).dt.days / 365.25

# Se filtran sólo menores de 15 años

df_pediatricos = df_trasplantes[df_trasplantes['EDAD'] < 15].copy()

# Se Verifican resultados

print("Registros totales de trasplante:", len(df_trasplantes))
print("Pacientes < 15 años:", len(df_pediatricos))
df_pediatricos.head()
```

```python
import numpy as np

# Definimos los grupos etáreos
bins = [0, 5, 10, 15]
labels = ['0-4 años', '5-9 años', '10-14 años']
df_pediatricos['Grupo Etario'] = pd.cut(df_pediatricos['EDAD'], bins=bins, labels=labels, right=False)

# Lista de códigos del top 10 global
top10_codigos = top10['Código diagnóstico'].tolist()
top10_nombres = top10['Nombre'].tolist()

# Creamos un DataFrame vacío para almacenar las frecuencias
datos_grupos = pd.DataFrame(index=top10_nombres, columns=labels)

# Calculamos la frecuencia por cada grupo etáreo
for grupo in labels:
    grupo_df = df_pediatricos[df_pediatricos['Grupo Etario'] == grupo]
    freq = (
        grupo_df[diagnosticos]
        .stack()
        .value_counts()
        .reindex(top10_codigos)
        .fillna(0)
        .values
    )
    datos_grupos[grupo] = freq

# Gráfico de barras agrupadas
x = np.arange(len(top10_nombres))
ancho_barra = 0.25

fig, ax = plt.subplots(figsize=(22, 9))

# Dibujar barras por grupo etáreo
for i, grupo in enumerate(labels):
    ax.bar(x + i * ancho_barra, datos_grupos[grupo], ancho_barra, label=grupo)

# Configuración del gráfico
ax.set_xlabel('Diagnóstico')
ax.set_ylabel('Frecuencia')
ax.set_title('Frecuencia de diagnósticos top 10 por grupo etáreo')
ax.set_xticks(x + ancho_barra)
ax.set_xticklabels(top10_nombres, rotation=45, ha='right')
ax.legend(title="Grupos Etáreos")

# Añadir etiquetas sobre las barras
for i, grupo in enumerate(labels):
    for j in x:
        ax.text(
            j + i * ancho_barra,
            datos_grupos[grupo].iloc[j] + 0.5,
            f'{int(datos_grupos[grupo].iloc[j])}',
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro

# Sumar los totales por año
evolucion_total = evolucion.sum(axis=1)

# Test de normalidad (Shapiro-Wilk)
stat, p_shapiro = shapiro(evolucion_total)

print(f"p-valor del test de Shapiro-Wilk: {p_shapiro:.4f}")

if p_shapiro > 0.05:
    print("Los datos tienen distribución normal. Se puede aplicar ANOVA.")

    # ANOVA de una vía
    from scipy.stats import f_oneway
    stat_anova, p_anova = f_oneway(
        evolucion.loc[2019],
        evolucion.loc[2020],
        evolucion.loc[2021],
        evolucion.loc[2022]
    )
    print(f"p-valor del ANOVA: {p_anova:.4f}")

    if p_anova < 0.05:
        print("Hay diferencias significativas entre los años (ANOVA).")
    else:
        print("No hay diferencias significativas entre los años (ANOVA).")

else:
    print("Los datos NO tienen distribución normal")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Asegura formatos de fecha
df_pediatricos["FECHA_EGRESO"] = pd.to_datetime(df_pediatricos["FECHA_EGRESO"], errors="coerce")
df_pediatricos["FECHA_INGRESO"] = pd.to_datetime(df_pediatricos["FECHA_INGRESO"], errors="coerce")

# Calcular delta_hp (días de hospitalización)
df_pediatricos["delta_hp"] = (df_pediatricos["FECHA_EGRESO"] - df_pediatricos["FECHA_INGRESO"]).dt.days

# Filtrar valores válidos
delta = df_pediatricos["delta_hp"].dropna()
delta = delta[delta.between(0, 365)]  # Solo primeros 100 días para claridad

# Conteo exacto
conteo_dias = delta.value_counts().sort_index()

# Colores diferenciados
colors = ['#FFA726' if d <= 30 else '#42A5F5' for d in conteo_dias.index]
labels = [f"Día {d}" for d in conteo_dias.index]

# Cálculo de porcentajes
total_pacientes = len(delta)
corto = delta[delta <= 30]
mediano = delta[(delta > 30) & (delta <= 365)]
corto_pct = len(corto) / total_pacientes * 100
mediano_pct = len(mediano) / total_pacientes * 100

# Graficar
plt.figure(figsize=(20, 6))
bars = plt.bar(conteo_dias.index, conteo_dias.values,
               color=colors, edgecolor='black', width=0.9)

# Etiquetas de día debajo de cada barra
plt.xticks(conteo_dias.index, labels, rotation=90, fontsize=9)

# Corte visual a 30 días
plt.axvline(30, color='red', linestyle='--', linewidth=1.2, label='Corte: 30 días')

# Caja con porcentajes
plt.text(0.75, 0.95,
         f"Corto plazo: {len(corto)} pacientes ({corto_pct:.1f}%)\n"
         f"Mediano plazo: {len(mediano)} pacientes ({mediano_pct:.1f}%)",
         transform=plt.gca().transAxes,
         fontsize=11,
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))


# Ejes y título
plt.xlabel("Días de hospitalización (delta h–p)", fontsize=12)
plt.ylabel("Número de pacientes", fontsize=12)
plt.title("Duración de hospitalización", fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
