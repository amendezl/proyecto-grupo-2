```python
 # Esta funcion busca inconsistencias entre registros con un mismo cip_encriptado
def detectar_cip_inconsistentes(df_todos, columnas_verificacion, columna_id='CIP_ENCRIPTADO'):
    # Agrupar por CIP y contar cuántos valores únicos hay en cada columna de verificación
    resumen_inconsistencias = df_todos.groupby(columna_id)[columnas_verificacion] \
        .nunique()

    # Encontrar los CIP que tienen más de 1 valor distinto en al menos una columna
    cip_inconsistentes = resumen_inconsistencias[
        (resumen_inconsistencias > 1).any(axis=1)
    ].index

    # Filtrar el DataFrame original para mostrar solo esos CIP
    df_inconsistentes = df_todos[df_todos[columna_id].isin(cip_inconsistentes)].copy()

    print(f"CIP con datos inconsistentes: {len(cip_inconsistentes)}")
    print(f"Registros involucrados: {df_inconsistentes.shape[0]}")

    return df_inconsistentes, cip_inconsistentes
```

```python
columnas_a_verificar = ['SEXO', 'FECHA_NACIMIENTO']

df_inconsistentes, cips_inconsistentes = detectar_cip_inconsistentes(df_todos, columnas_a_verificar)
```

```python
# Se eliminan esos registros con inconsistencias
antes = df_todos.shape[0]
df_todos = df_todos[~df_todos['CIP_ENCRIPTADO'].isin(cips_inconsistentes)].copy()
despues = df_todos.shape[0]
print(f"Registros eliminados por CIP inconsistente: {antes - despues}")
```

```python
# Se definen listas de columnas de procedimientos y diagnósticos

procedimientos = [f'PROCEDIMIENTO{i}' for i in range(1, 21)]
diagnosticos   = [f'DIAGNOSTICO{i}'  for i in range(1, 21)]

# Se asegura de que cada código está como string sin espacios ni comas

for col in procedimientos:
    df_todos[col] = df_todos[col]\
        .astype(str)\
        .str.replace(',', '.', regex=False)\
        .str.strip()
```

```python
import pandas as pd


# 1) Filas que contienen trasplante en alguna columna de procedimientos
mask_trasplante = df_todos[procedimientos].isin({"55.6","55.61","55.69"}).any(axis=1)

# 2) Fecha del primer trasplante por paciente
primer_tx = (
    df_todos.loc[mask_trasplante, ["CIP_ENCRIPTADO", "FECHA_INGRESO"]]
            .groupby("CIP_ENCRIPTADO", as_index=False)["FECHA_INGRESO"].min()
            .rename(columns={"FECHA_INGRESO": "FECHA_PRIMER_TRASPLANTE"})
)

# 3) Unir sin tocar df_todos y filtrar trasplante + posteriores (incluye el día del trasplante)
df_trasplante_y_post = (
    df_todos.merge(primer_tx, on="CIP_ENCRIPTADO", how="inner")
            .loc[lambda d: d["FECHA_INGRESO"] >= d["FECHA_PRIMER_TRASPLANTE"]]
            .sort_values(["CIP_ENCRIPTADO", "FECHA_INGRESO"])
            .reset_index(drop=True)
)

print("Pacientes con ≥1 trasplante:", primer_tx["CIP_ENCRIPTADO"].nunique())
print("Registros (trasplante + posteriores):", len(df_trasplante_y_post))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, f_oneway, kruskal
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Utilidades y preparación
diagnosticos = [f'DIAGNOSTICO{i}' for i in range(1, 21)]
no_validos = {'', 'NAN', 'SIN INFORMACIÓN', 'DESCONOCIDO'}

df_tmp = df_pediatricos.copy()
df_tmp['FECHA_INGRESO'] = pd.to_datetime(df_tmp['FECHA_INGRESO'], errors='coerce')

# 2) Top-6 global por PACIENTES ÚNICOS (sin año)
#    (1 paciente cuenta 1 vez por código)
df_diag_global = (
    df_tmp[['CIP_ENCRIPTADO'] + diagnosticos]
      .melt(id_vars='CIP_ENCRIPTADO', value_vars=diagnosticos,
            var_name='Posicion', value_name='Codigo')
)
df_diag_global['Codigo'] = (
    df_diag_global['Codigo'].astype(str).str.strip().str.upper()
)

df_diag_global = df_diag_global[
    df_diag_global['Codigo'].notna() & ~df_diag_global['Codigo'].isin(no_validos)
].drop_duplicates(['CIP_ENCRIPTADO', 'Codigo'])

top6_codigos = (
    df_diag_global.groupby('Codigo')['CIP_ENCRIPTADO']
    .nunique()
    .sort_values(ascending=False)
    .head(6)
    .index.tolist()
)

# 3) Año del PRIMER trasplante y diagnósticos de dicha hospitalización (1 fila por paciente)
df_primer = (
    df_tmp.sort_values('FECHA_INGRESO')
          .drop_duplicates('CIP_ENCRIPTADO', keep='first')
          .loc[:, ['CIP_ENCRIPTADO', 'FECHA_INGRESO'] + diagnosticos]
          .assign(Año=lambda d: d['FECHA_INGRESO'].dt.year)
)

# Diagnósticos de la primera hospitalización en formato largo
df_diag_first = (
    df_primer.melt(
        id_vars=['CIP_ENCRIPTADO', 'Año'],
        value_vars=diagnosticos,
        var_name='Posicion', value_name='Codigo'
    )
)
df_diag_first['Codigo'] = (
    df_diag_first['Codigo'].astype(str).str.strip().str.upper()
)

df_diag_first = df_diag_first[
    df_diag_first['Codigo'].notna() & ~df_diag_first['Codigo'].isin(no_validos)
].drop_duplicates(['CIP_ENCRIPTADO', 'Codigo'])

# Filtrar a Top6 para estabilizar columnas
df_diag_first = df_diag_first[df_diag_first['Codigo'].isin(top6_codigos)]

# 4) EVOLUCIÓN (Año × Código) en PACIENTES ÚNICOS (N)
evolucion = pd.pivot_table(
    df_diag_first,
    index='Año', columns='Codigo',
    values='CIP_ENCRIPTADO',
    aggfunc='nunique', fill_value=0
).astype(int)

evolucion = (
    evolucion
    .reindex(index=[2019, 2020, 2021, 2022, 2023])
    .reindex(columns=top6_codigos, fill_value=0)
    .fillna(0).astype(int)
)

# N pacientes usados (primer trasplante)
N = df_primer['CIP_ENCRIPTADO'].nunique()
print(f"N pacientes únicos (primer trasplante): {N}")
print("Matriz 'evolucion' (Año × Top6, pacientes únicos):\n", evolucion)

# 5) Estadística: Shapiro y ANOVA/Kruskal
evolucion_total = evolucion.sum(axis=1)
evolucion_total = evolucion_total.dropna()

if len(evolucion_total) < 3:
    raise ValueError("Se requieren ≥3 años para aplicar Shapiro/ANOVA.")

stat_sw, p_shapiro = shapiro(evolucion_total)
print(f"\nShapiro-Wilk sobre totales por año: p = {p_shapiro:.4f}")

if p_shapiro > 0.05:
    print("Distribución ~normal. Aplico ANOVA de una vía.")
    grupos = [evolucion.loc[anio].values for anio in evolucion.index.dropna()]
    stat_anova, p_anova = f_oneway(*grupos)
    print(f"ANOVA: p = {p_anova:.4f}")
    if p_anova < 0.05:
        print("Diferencias significativas entre años (ANOVA).")
    else:
        print("No hay diferencias significativas entre años (ANOVA).")
else:
    print("Datos NO normales. Aplico prueba de Kruskal–Wallis.")
    grupos = [evolucion.loc[anio].values for anio in evolucion.index.dropna()]
    stat_kw, p_kw = kruskal(*grupos)
    print(f"Kruskal–Wallis: p = {p_kw:.4f}")
    if p_kw < 0.05:
        print("Diferencias significativas entre años (Kruskal–Wallis).")
    else:
        print("No hay diferencias significativas entre años (Kruskal–Wallis).")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 0) Parámetros
diagnosticos = [f'DIAGNOSTICO{i}' for i in range(1, 21)]
no_validos = {'', 'NAN', 'SIN INFORMACIÓN', 'DESCONOCIDO'}
paleta_plazos = {"Corto plazo": "#66BB6A", "Mediano plazo": "#FFA726", "Largo plazo": "#42A5F5"}
orden_plazos = ["Corto plazo", "Mediano plazo", "Largo plazo"]
codigos_permitidos = list(nombres_diagnosticos.keys())
nombres_permitidos = [nombres_diagnosticos[c] for c in codigos_permitidos]

# 1) Plazo por paciente (1ª hospitalización)
d = df_pediatricos.copy()
d["FECHA_INGRESO"] = pd.to_datetime(d["FECHA_INGRESO"], errors="coerce")
d["FECHA_EGRESO"]  = pd.to_datetime(d["FECHA_EGRESO"],  errors="coerce")
d["DIAS_ESTANCIA"] = (d["FECHA_EGRESO"] - d["FECHA_INGRESO"]).dt.days
d = d.dropna(subset=["DIAS_ESTANCIA"]).loc[d["DIAS_ESTANCIA"].between(0, 365)]

df_first = (
    d.sort_values("FECHA_INGRESO")
     .drop_duplicates("CIP_ENCRIPTADO", keep="first")
     .copy()
)

# Regla: μ dentro de ≤30 días
corto_pool = df_first.loc[df_first["DIAS_ESTANCIA"] <= 30, "DIAS_ESTANCIA"]
mu_corto = 30.0 if corto_pool.empty else corto_pool.mean()

def asignar_plazo(x):
    if x < mu_corto:   return "Corto plazo"
    elif x <= 30:      return "Mediano plazo"
    else:              return "Largo plazo"

df_first["Plazo"] = df_first["DIAS_ESTANCIA"].apply(asignar_plazo)

# 2) Paciente–código únicos (solo los permitidos)
df_diag = (
    df_pediatricos[['CIP_ENCRIPTADO'] + diagnosticos]
      .melt(id_vars='CIP_ENCRIPTADO', value_vars=diagnosticos,
            var_name='Posicion', value_name='Codigo')
)
df_diag['Codigo'] = df_diag['Codigo'].astype(str).str.strip().str.upper()

df_pac_cod = (df_diag[df_diag['Codigo'].notna() &
                      ~df_diag['Codigo'].isin(no_validos) &
                      df_diag['Codigo'].isin(codigos_permitidos)]
              .drop_duplicates(['CIP_ENCRIPTADO', 'Codigo']))

# Anexar el plazo (definido por la 1ª hospitalización)
df_pac_cod = df_pac_cod.merge(df_first[['CIP_ENCRIPTADO', 'Plazo']],
                              on='CIP_ENCRIPTADO', how='inner')

# 3) Tabla: pacientes únicos por Código (solo permitidos) × Plazo
tabla = (df_pac_cod
         .groupby(['Codigo','Plazo'])['CIP_ENCRIPTADO'].nunique()
         .unstack('Plazo')
         .reindex(index=codigos_permitidos)
         .reindex(columns=orden_plazos)
         .fillna(0).astype(int))

# 4) Barras agrupadas (3 por diagnóstico)
x = np.arange(len(tabla.index))
ancho = 0.25

fig, ax = plt.subplots(figsize=(20, 9))
for i, plazo in enumerate(orden_plazos):
    vals = tabla[plazo].values
    ax.bar(x + i*ancho, vals, width=ancho,
           color=paleta_plazos[plazo], edgecolor='black', label=plazo)
    # etiquetas
    for j, v in enumerate(vals):
        ax.text(x[j] + i*ancho, v + 0.4, int(v),
                ha='center', va='bottom', fontsize=9)

# Etiquetas del eje X con los nombres del diccionario
etiquetas = [nombres_diagnosticos[c] for c in tabla.index]
ax.set_xticks(x + ancho)
ax.set_xticklabels(etiquetas, rotation=45, ha='right')

ax.set_xlabel('Diagnóstico')
ax.set_ylabel('Pacientes únicos')
ax.set_title('Diagnósticos seleccionados por plazo (pacientes únicos, primera hospitalización)')
ax.legend(title='Plazo')
ax.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
print("Códigos graficados (en orden):", list(tabla.index))
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from pathlib import Path

# 1) Preparación del dataset
# a) Primera hospitalización post-trasplante por paciente
d = df_pediatricos.copy()
d["FECHA_INGRESO"] = pd.to_datetime(d["FECHA_INGRESO"], errors="coerce")
d["FECHA_NACIMIENTO"] = pd.to_datetime(d["FECHA_NACIMIENTO"], errors="coerce")

df_first = (
    d.sort_values(["CIP_ENCRIPTADO", "FECHA_INGRESO"])
     .drop_duplicates("CIP_ENCRIPTADO", keep="first")
     .copy()
)

# b) Edad al primer contacto post-trasplante (en años) y grupos etarios
df_first["EDAD_TX"] = (df_first["FECHA_INGRESO"] - df_first["FECHA_NACIMIENTO"]).dt.days / 365.25
bins = [0, 6.0000001, 9.0000001, 12.0000001, 14.9999999]
labels = ['0-6', '7-9', '10-12', '13-14']
df_first["GRUPO_ETARIO"] = pd.cut(df_first["EDAD_TX"].clip(lower=0, upper=14.999), bins=bins,
                                  labels=labels, include_lowest=True, right=True)

# c) Diagnósticos a considerar (DIAGNOSTICO1–DIAGNOSTICO20) y Top-6
diag_cols = [c for c in [f"DIAGNOSTICO{i}" for i in range(1, 21)] if c in df_first.columns]
if len(diag_cols) == 0:
    raise ValueError("No se encontraron columnas DIAGNOSTICO1–DIAGNOSTICO20.")

# Top-6 definidos en el notebook
codigos_top6 = ['I15.1', 'N39.0', 'Y83.0', 'E46', 'N25.8', 'T86.1']
no_validos = {'', 'NAN', 'SIN INFORMACIÓN', 'DESCONOCIDO', None, np.nan}

def _prepara_codigos(row):
    cods = []
    for c in diag_cols:
        v = row.get(c)
        if pd.isna(v):
            continue
        v = str(v).strip()
        if v in no_validos:
            continue
        cods.append(v)
    return cods

# d) Variable dependiente Y:
def cuenta_top6_distintos(cods):
    return len(set([c for c in cods if c in codigos_top6]))

def cuenta_top6_ocurrencias(cods):
    return sum(1 for c in cods if c in codigos_top6)

codigos_list = df_first.apply(_prepara_codigos, axis=1)
df_first["N_TOP6_DISTINTOS"]   = codigos_list.apply(cuenta_top6_distintos)
df_first["N_TOP6_OCURRENCIAS"] = codigos_list.apply(cuenta_top6_ocurrencias)

TARGET = "N_TOP6_DISTINTOS"

# e) Variables categóricas: SEXO y REGIÓN
region_col = "REGIÓN" if "REGIÓN" in df_first.columns else ("REGION" if "REGION" in df_first.columns else None)
if region_col is None:
    raise KeyError("No se encontró columna de región ('REGIÓN' o 'REGION').")

df_modelo = (
    df_first[["CIP_ENCRIPTADO", "SEXO", "GRUPO_ETARIO", region_col, TARGET]]
    .rename(columns={region_col: "REGION"})
    .dropna(subset=["SEXO", "GRUPO_ETARIO", "REGION", TARGET])
    .copy()
)

df_modelo["SEXO"] = df_modelo["SEXO"].astype("category")
df_modelo["GRUPO_ETARIO"] = df_modelo["GRUPO_ETARIO"].astype("category")
df_modelo["REGION"] = df_modelo["REGION"].astype("category")
ref_sexo = "MUJER" if "MUJER" in df_modelo["SEXO"].cat.categories else df_modelo["SEXO"].cat.categories[0]
ref_ge   = "0-6"   if "0-6"   in df_modelo["GRUPO_ETARIO"].cat.categories else df_modelo["GRUPO_ETARIO"].cat.categories[0]
ref_reg  = "METROPOLITANA" if "METROPOLITANA" in df_modelo["REGION"].cat.categories else df_modelo["REGION"].cat.categories[0]
```

```python
# 2) Ajuste del modelo
formula = (
    f'{TARGET} ~ '
    f'C(SEXO, Treatment(reference="{ref_sexo}")) + '
    f'C(GRUPO_ETARIO, Treatment(reference="{ref_ge}")) + '
    f'C(REGION, Treatment(reference="{ref_reg}"))'
)

modelo_ols = smf.ols(formula, data=df_modelo).fit(cov_type="HC1")

print("\n==============================================")
print("Regresión lineal multivariada (OLS) — Coeficientes")
print("Y =", TARGET, "| Referencias -> SEXO:", ref_sexo, "| GRUPO_ETARIO:", ref_ge, "| REGION:", ref_reg)
print("==============================================\n")
print(modelo_ols.summary())

# 3) Tabla de coeficientes limpia + IC95%
coef_table = pd.DataFrame({
    "Termino": modelo_ols.params.index,
    "Coeficiente": modelo_ols.params.values,
    "EE robusta (HC1)": modelo_ols.bse.values,
    "t": modelo_ols.tvalues.values,
    "p-valor": modelo_ols.pvalues.values
})
ci = modelo_ols.conf_int(alpha=0.05)
coef_table["IC95% inferior"] = ci[0].values
coef_table["IC95% superior"] = ci[1].values

# Orden: Intercept primero, luego covariables
def _orden(t):
    if t == "Intercept":
        return (0, "")
    if t.startswith("C(SEXO"):
        return (1, t)
    if t.startswith("C(GRUPO_ETARIO"):
        return (2, t)
    if t.startswith("C(REGION"):
        return (3, t)
    return (9, t)

coef_table = coef_table.sort_values(by="Termino", key=lambda s: s.map(_orden)).reset_index(drop=True)

display(coef_table)
```
