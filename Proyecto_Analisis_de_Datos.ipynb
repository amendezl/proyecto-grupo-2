```python
# Se definen listas de columnas de procedimientos y diagnósticos

procedimientos = [f'PROCEDIMIENTO{i}' for i in range(1, 11)]
diagnosticos   = [f'DIAGNOSTICO{i}'  for i in range(1, 11)]

# Se asegura de que cada código está como string sin espacios ni comas

for col in procedimientos:
    df_todos[col] = df_todos[col]\
        .astype(str)\
        .str.replace(',', '.', regex=False)\
        .str.strip()
```

```python
# Se define el set de códigos de trasplante renal

codigos_trasplante = {'55.6', '55.61', '55.69'}

# Se construye la máscara: True si alguna de las 10 columnas coincide
mask_trasplante = df_todos[procedimientos] \
    .isin(codigos_trasplante) \
    .any(axis=1)

# Se aplica el filtro
df_trasplantes = df_todos[mask_trasplante].copy()

# 4. Verificar
print("Registros con trasplante renal:", df_trasplantes.shape[0])

df_trasplantes.head()
```

```python
# Se Calcula la edad en años a la fecha de ingreso

df_trasplantes['EDAD'] = (
    df_trasplantes['FECHA_INGRESO'] - df_trasplantes['FECHA_NACIMIENTO']
).dt.days / 365.25

# Se filtran sólo menores de 15 años

df_pediatricos = df_trasplantes[df_trasplantes['EDAD'] < 15].copy()

# Se Verifican resultados

print("Registros totales de trasplante:", len(df_trasplantes))
print("Pacientes < 15 años:", len(df_pediatricos))
df_pediatricos.head()
```

```python
# Se usan solo las columnas desde DIAGNOSTICO2 hasta DIAGNOSTICO10 para marcar complicaciones

secundarios = diagnosticos[1:]

# Se crea la columna de complicación (1 si tiene, 0 si no tiene)

df_pediatricos['TIENE_COMPLICACION'] = (
    df_pediatricos[secundarios]
    .notnull()
    .any(axis=1)
).astype(int)

print("Total pediátricos:", len(df_pediatricos))
print("Con complicación:", df_pediatricos['TIENE_COMPLICACION'].sum())
print("Sin complicación:", (df_pediatricos['TIENE_COMPLICACION']==0).sum())
print("\n% con complicación: {:.2%}".format(df_pediatricos['TIENE_COMPLICACION'].mean()))

df_pediatricos[['CIP_ENCRIPTADO', 'EDAD', *secundarios, 'TIENE_COMPLICACION']].head()
```

```python
# Se apilan los valores de esas columnas en una sola serie

codigos_apilados = df_pediatricos[secundarios].stack()

# Se crea una serie y se cuentan frecuencias

serie_codigos = pd.Series(codigos_apilados.values, name='Código diagnóstico')
frecuencia = serie_codigos.value_counts().reset_index()
frecuencia.columns = ['Código diagnóstico', 'Frecuencia']

print(frecuencia)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Función para filtrar outliers por IQR dentro de cada grupo ---
def filtrar_outliers_iqr(df, columna, grupos):
    """
    Elimina las filas cuya columna esté fuera de [Q1 - 1.5·IQR, Q3 + 1.5·IQR]
    calculado dentro de cada grupo.
    """
    def mask_iqr(x):
        q1 = x.quantile(0.25)
        q3 = x.quantile(0.75)
        iqr = q3 - q1
        lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
        return x.between(lower, upper)

    mask = df.groupby(grupos)[columna].transform(mask_iqr)
    return df[mask]

# --- 2. Filtrar outliers ---
# Asume que tu DataFrame original se llama df_plot
grupos = ['DIAGNOSTICO2', 'SEXO']
df_limpio = filtrar_outliers_iqr(df_plot, 'DIAS_ESTANCIA', grupos)

# --- 3. Crear copia para no tener SettingWithCopyWarning ---
df_limpio = df_limpio.copy()

# --- 4. Mapear códigos a sus nombres verdaderos ---
nombres_diagnosticos = {
    'D59.3': 'Síndrome hemolítico-urémico',
    'D72.8': 'Otros trastornos especificados de los leucocitos',
    'E83.4': 'Trastornos del metabolismo del magnesio',
    'H50.9': 'Estrabismo, no especificado',
    'I15.1': 'Hipertensión secundaria a otros trastornos renales',
    'I77.8': 'Otros trastornos especificados de arterias y arteriolas',
    'K91.8': 'Otros trastornos del sistema digestivo consecutivos a procedimientos, no clasificados en otra parte',
    'M31.1': 'Microangiopatía trombótica',
    'M32.1': 'Lupus eritematoso sistémico con compromiso de órganos o sistemas',
    'N00.5': 'Síndrome nefrítico agudo, glomerulonefritis mesangiocapilar difusa',
    'N02.8': 'Hematuria recurrente y persistente, otras',
    'N03.5': 'Síndrome nefrítico crónico, glomerulonefritis mesangiocapilar difusa',
    'N04.1': 'Síndrome nefrótico, lesiones glomerulares focales y segmentarias',
    'N04.3': 'Síndrome nefrótico, glomerulonefritis proliferativa mesangial difusa',
    'N04.8': 'Síndrome nefrótico, otras',
    'N04.9': 'Síndrome nefrótico, no especificada',
    'N07.1': 'Nefropatía hereditaria, no clasificada en otra parte, lesiones glomerulares focales y segmentarias',
    'N08.5': 'Trastornos glomerulares en trastornos sistémicos del tejido conjuntivo',
    'N18.5': 'Enfermedad renal crónica, estadio 5',
    'N39.0': 'Infección de vías urinarias, sitio no especificado',
    'Q53.1': 'Testículo no descendido, unilateral',
    'Q61.1': 'Riñón poliquístico, autosómico recesivo',
    'Q61.3': 'Riñón poliquístico, tipo no especificado',
    'Q61.4': 'Displasia renal',
    'Q61.5': 'Riñón quístico medular',
    'Q61.8': 'Otras enfermedades renales quísticas',
    'Q62.3': 'Otros defectos obstructivos de la pelvis renal y del uréter',
    'Q63.1': 'Riñón lobulado, fusionado y en herradura',
    'Q64.2': 'Válvulas uretrales posteriores congénitas',
    'Q79.4': 'Síndrome del abdomen en ciruela pasa',
    'T50.9': 'Otras drogas y sustancias biológicas, y las no especificadas',
    'T81.0': 'Hemorragia y hematoma que complican un procedimiento, no clasificados en otra parte',
    'T86.1': 'Falla y rechazo de trasplante de riñón',
    'Y83.0': 'Operación quirúrgica con trasplante de un órgano completo',
    'Z49.2': 'Otras diálisis',
    'Z90.5': 'Ausencia adquirida de riñón',
    'Z99.2': 'Dependencia de diálisis renal'
}

df_limpio['Nombre_Diagnostico'] = df_limpio['DIAGNOSTICO2'].map(nombres_diagnosticos)

# --- 5. Dibujar el boxplot usando el nombre en el eje X ---
plt.figure(figsize=(15, 8))
sns.boxplot(
    data=df_limpio,
    x='Nombre_Diagnostico',
    y='DIAS_ESTANCIA',
    hue='SEXO',
    showfliers=False
)
plt.title('Estancia hospitalaria por diagnóstico y sexo')
plt.xlabel('Diagnóstico')
plt.ylabel('Días de hospitalización')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import mannwhitneyu
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1) Función de test con cálculo de U, Z y r ---
def test_mannwhitney(muestra1, muestra2, nombre1='Grupo 1', nombre2='Grupo 2'):
    U, p_val = mannwhitneyu(muestra1, muestra2, alternative='two-sided')
    n1, n2 = len(muestra1), len(muestra2)
    mean_U = n1 * n2 / 2
    std_U  = np.sqrt(n1 * n2 * (n1 + n2 + 1) / 12)
    z = (U - mean_U) / std_U
    r = abs(z) / np.sqrt(n1 + n2)

    print(f"Mann–Whitney U ({nombre1} vs {nombre2})")
    print(f"  U = {U:.2f}")
    print(f"  p-valor = {p_val:.4f}")
    print(f"  Z = {z:.3f}")
    print(f"  Efecto-tamaño r = {r:.3f}")
    alpha = 0.05
    if p_val < alpha:
        print(f"→ p < {alpha}: diferencias SIGNIFICATIVAS.")
    else:
        print(f"→ p ≥ {alpha}: no hay evidencia de diferencia.")

# --- 2) Extraer muestras de días de estancia ---
mujer  = df_pediatricos.loc[df_pediatricos['SEXO']=='MUJER',  'DIAS_ESTANCIA']
hombre = df_pediatricos.loc[df_pediatricos['SEXO']=='HOMBRE','DIAS_ESTANCIA']

# --- 3) Ejecutar y mostrar resultados por consola ---
test_mannwhitney(mujer, hombre, 'Mujeres', 'Hombres')

# --- 4) Dibujar boxplot y anotar p-valor ---
fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(
    data=df_pediatricos,
    x='SEXO',
    y='DIAS_ESTANCIA',
    showfliers=False,
    ax=ax
)
ax.set(
    title='Días de estancia por sexo (pediátricos)',
    xlabel='',
    ylabel='Días de hospitalización'
)

# Obtener el p-valor de nuevo para anotar
U, p_val = mannwhitneyu(mujer, hombre, alternative='two-sided')
txt = f"p = {p_val:.3f}"

# Calcular posición: un poco por debajo del tope del eje Y
ymin, ymax = ax.get_ylim()
ax.text(
    0.5,
    ymax - (ymax - ymin)*0.05,
    txt,
    ha='center',
    va='top',
    fontsize=12
)

# --- 5) Ajuste manual de márgenes para que el plot ocupe bien el área ---
fig.subplots_adjust(
    left=0.12,   # espacio izquierdo
    right=0.95,  # espacio derecho
    top=0.90,    # espacio superior
    bottom=0.12  # espacio inferior
)

plt.show()
```
